```python
#Create function to pull the csv based on name - btw Beautiful Soup will not work
import urllib.request
import urllib.parse
def pull(l): 
    url = "http://cycling.data.tfl.gov.uk/usage-stats/{}".format(urllib.parse.quote(l[:]))
    print(url)
    urllib.request.urlretrieve(url, l[:])
    print('Done')
```

```python
#Use Tinker to select target folder - IMPORTANT NOT TO SHARE LOCATION ON GITHUB :) 
from tkinter import filedialog
from tkinter import *
root = Tk()
root.withdraw()
folder_selected = filedialog.askdirectory()
```

```python
#"*2021.cs"
def pullbatch(searchstring):
    all_files = glob.glob(os.path.join(folder_selected , searchstring))
    li = []
    for i, filename in enumerate(all_files):
        print("--> Reading ", i , " csv file" )
        df = pd.read_csv(filename, index_col=None, header=0)
        df['filename'] = filename
        df['filename'] = df['filename'].str.replace(folder_selected,'')
        #df['Duration(Min)'] = df['Duration'] / 60.00
        li.append(df)
        print("Done")
    bigdf = pd.concat(li, axis=0, ignore_index=True)
    bigdf.info()
    return bigdf
df2017 = pullbatch("*17.cs")
```

```python
df2017['EndStation Idx'] = df2017['EndStation Id']
df2017.loc[df2017['EndStation Idx'].isnull(),'EndStation Idx' ] = df2017['End Station Id']
df2017['EndStation Namex'] = df2017['End Station Name']
df2017.loc[df2017['EndStation Namex'].isnull(),'EndStation Namex' ] = df2017['End Station Name']
df2017['StartStation Idx'] = df2017['StartStation Id']
df2017.loc[df2017['StartStation Idx'].isnull(),'StartStation Idx' ] = df2017['Start Station Id']
df2017['StartStation Namex'] = df2017['Start Station Name']
df2017.loc[df2017['StartStation Namex'].isnull(),'StartStation Namex' ] = df2017['Start Station Name']
```

```python
df2017.columns
```

```python
#Tidy Up Data
df2017['EndStation Id'] = df2017['EndStation Idx']
df2017['End Station Name'] = df2017['EndStation Namex']
df2017['StartStation Id'] = df2017['StartStation Idx'] 
df2017['Start Station Name'] = df2017['StartStation Namex']
df2017 = df2017[['Rental Id', 'Duration', 'Bike Id', 'End Date', 'EndStation Id',
       'EndStation Name', 'Start Date', 'StartStation Id', 'StartStation Name',
       'filename', 'Duration_Seconds']]
df2017.head()
```

```python
df2017agg = df2017.groupby(['End Date', 'EndStation Id','EndStation Name', 'Start Date', 'StartStation Id', 'StartStation Name'],dropna=False).size().reset_index()
```

```python
#"*2021.cs"
def pullbatchfull(searchstring,file1name, file2name):
    all_files = glob.glob(os.path.join(folder_selected , searchstring))
    li = []
    print("--> Reading csv files for ", searchstring )
    for i, filename in enumerate(all_files):
        df = pd.read_csv(filename, index_col=None, header=0)
        df['filename'] = filename
        df['filename'] = df['filename'].str.replace(folder_selected,'')
        #df['Duration(Min)'] = df['Duration'] / 60.00
        li.append(df)
    print("Done")
    bigdf = pd.concat(li, axis=0, ignore_index=True)
    bigdfagg = bigdf.groupby(['End Date', 'EndStation Id','EndStation Name', 'Start Date', 'StartStation Id', 'StartStation Name'],dropna=False).size().reset_index()
    print("Fulldata size: " ,bigdf.shape)
    print("Rollupdata size: " ,bigdfagg.shape)
    print("Export to csv start")
    bigdf.to_csv(file1name, index = False)
    bigdfagg.to_csv(file2name, index = False )
    print("Export to csv complete")
    return bigdf , bigdfagg
#df2018 , df2018agg = pullbatchfull("*18.cs" , 'TFLBike2018.csv' , 'TFLBike2018Aggregated.csv')
#df2019 , df2019agg = pullbatchfull("*2019.cs" , 'TFLBike2019.csv' , 'TFLBike2019Aggregated.csv')
df2020 , df2020agg = pullbatchfull("*2020.cs" , 'TFLBike2020.csv' , 'TFLBike2020Aggregated.csv')
df2021 , df2021agg = pullbatchfull("*2021.cs" , 'TFLBike2021.csv' , 'TFLBike2021Aggregated.csv')
```
